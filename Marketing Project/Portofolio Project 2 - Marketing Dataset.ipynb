#Libraries Needed

#Cleaning Data

```python
df.info()
```

```python
df.isna().sum()
```

```python
df[df[' Income '].isna()]
```

```python
df[' Income '] = df[' Income '].fillna(df[' Income '].median())
df[' Income '].isna().any()
```

```python
df.duplicated().values.any()
```

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])
df['Dt_Customer']
```

```python
marital_status = df['Marital_Status'].value_counts()
marital_status
```

```python
df[(df['Marital_Status'] == 'Absurd') | (df['Marital_Status'] == 'YOLO')]
```

```python
df = df.drop(df[df['Marital_Status'] == 'YOLO'].index)
df = df.drop(df[df['Marital_Status'] == 'Absurd'].index)
```

#Data Analysis and Visualization

## 1. Know the age group of the customers.

```python
df[df['Year_Birth'] < 1940]
```

```python
#Due to Customer Date These People Were over 100 Years When They First Purchased, and That Doesn't Make any Sense
df = df.drop(df[df['Year_Birth'] < 1940].index)
```

```python
age_groups = df['Year_Birth'].value_counts()
age_groups
```

```python
fig = plt.figure(figsize=(10,8))
plt.bar(x=age_groups.index, height=age_groups.values, color='deeppink', width=0.8)
plt.show()
```

##2.   Know the customers education and which group buyes from you more.

```python
education_status = df['Education'].value_counts()
education_status
```

```python
fig = px.pie(df, labels=education_status.index, values=education_status.values,
             color_discrete_sequence= px.colors.qualitative.Set3, names=education_status.index)
fig.update_layout()
fig.show()
```

##6.   Know which product has the most sales and which is not.

```python
# Outlier Detection
df[df['MntMeatProducts'] == df['MntMeatProducts'].max()]
```

```python
fig = go.Figure()

fig.add_trace(go.Box(y=df['MntWines'], name='wines', marker=dict(color=px.colors.qualitative.Set3[0])))
fig.add_trace(go.Box(y=df['MntFruits'], name='Fruits', marker=dict(color=px.colors.qualitative.Set3[1])))
fig.add_trace(go.Box(y=df['MntMeatProducts'], name='Meat', marker=dict(color=px.colors.qualitative.Set3[2])))
fig.add_trace(go.Box(y=df['MntFishProducts'], name='Fish', marker=dict(color=px.colors.qualitative.Set3[3])))
fig.add_trace(go.Box(y=df['MntSweetProducts'], name='Sweet', marker=dict(color=px.colors.qualitative.Set3[4])))
fig.add_trace(go.Box(y=df['MntGoldProds'], name='Gold', marker=dict(color=px.colors.qualitative.Set3[5])))

fig.show()
```

##7.   Know which channel has the more or less purchases.

```python
fig = go.Figure()
fig.add_trace(go.Histogram(x=df['NumCatalogPurchases'], name='Catalog', marker=dict(color=px.colors.qualitative.Set3[0])))
fig.add_trace(go.Histogram(x=df['NumDealsPurchases'], name='Deals', marker=dict(color=px.colors.qualitative.Set3[1])))
fig.add_trace(go.Histogram(x=df['NumStorePurchases'], name='Store', marker=dict(color=px.colors.qualitative.Set3[2])))
fig.add_trace(go.Histogram(x=df['NumWebPurchases'], name='Web', marker=dict(color=px.colors.qualitative.Set3[3])))

# Overlay both histograms
fig.update_layout(barmode='overlay')
fig.show()
```

##8.   Know the average visits to your website.

```python
moving_avg_website = df['NumWebVisitsMonth'].rolling(window=3).mean()
```

```python
fig = px.scatter(df, x=moving_avg_website, y='Dt_Customer',
                   color=moving_avg_website)
fig.show()
```
